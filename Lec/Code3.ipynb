## SORU 7)

#### Kitabın ikinci bölümünün sonundaki 34. problemde verilen frekans cevabının genliğini ve fazını çiziniz.

```python
# Soruda verilen frekans cevabı: 
# H(e^jω) = (1-1.25e^(-jω))/(1-0.8e^(-jω)) veya
# H(e^jω) = 1-(0.45e^(-jω))/(1-0.8e^(-jω)) şeklindedir.

# ω'değerimizi tanımlayalım:
w=np.linspace(-pi,pi);
# frekans cevabımızı tanımlayalım:
H=(1-(0.45*np.exp(-1j*w))/(1-0.8*np.exp(-1j*w)));

# istenilen grafikleri çizelim.
amplitude=np.abs(H);  #abs() komutu ile genliğimizi,
phase=np.angle(H);    #angle() komutu ile fazımızı buluruz.

plt.figure(701);          # yeni figür açtık.
plt.subplot(2, 1, 1);     # subplot ile tek seferde yazdıralım.
plt.plot(w,amplitude);    # omegaya göre genlik çiz.
plt.title('$|H(e^{j\omega})|$'); # genlik için başlık.
plt.xlabel('ω');          # x ekseni ismi.
plt.grid();               # gridi aç.
print()
plt.subplot(2, 1, 2);     # subplot ile tek seferde yazdıralım.
plt.plot(w,phase);        # omegaya göre genlik çiz.
plt.title('$ \measuredangle H(e^{j\omega})$'); # genlik için başlık.
plt.xlabel('ω');          # x ekseni ismi.
plt.grid();               # gridi aç.

plt.show();               # grafiği göster.
plt.tight_layout();       # başlıklar çakışmasın diye.
```

#### Kitabın ikinci bölümünün sonundaki 32. problemde verilen frekans cevabının genliğini ve fazını 𝑎=−0.5 için çiziniz.

```python
# Soruda verilen frekans cevabı: 
# H(e^jω) = (1-j*e^(-jω)*)(1+j*e^(-jω)*)/(1-0.8e^(-jω)) veya
# H(e^jω) = (1+e^(-j2ω))/(1-0.8e^(-jω)) şeklindedir.

# frekans cevabımızı tanımlayalım:
H=(1+np.exp(-1j*2*w))/(1-0.8*np.exp(-1j*w));

# istenilen grafikleri çizelim.
amplitude=np.abs(H);  #abs() komutu ile genliğimizi,
phase=np.angle(H);    #angle() komutu ile fazımızı buluruz.

plt.figure(901);          # yeni figür açtık.
plt.subplot(2, 1, 1);     # subplot ile tek seferde yazdıralım.
plt.plot(w,amplitude);    # omegaya göre genlik çiz.
plt.title('$|H(e^{j\omega})|$'); # genlik için başlık.
plt.xlabel('ω');          # x ekseni ismi.
plt.grid();               # gridi aç.
print()
plt.subplot(2, 1, 2);     # subplot ile tek seferde yazdıralım.
plt.plot(w,phase);        # omegaya göre genlik çiz.
plt.title('$ \measuredangle H(e^{j\omega})$'); # genlik için başlık.
plt.xlabel('ω');          # x ekseni ismi.
plt.grid();               # gridi aç.

plt.show();               # grafiği göster.
plt.tight_layout();       # başlıklar çakışmasın diye.
```
