## KISIM 6)

```python
#soru 2.7 # a) şıkkı ......................

#sayımız r*e^jQ şeklindedir, (Q = teta için), N periyodunu N=12 Örnek bulmuştuk...

r =1;                                             # r yi 1 olarak tanımla.
Q =pi/6;                                          # soruda verilen teta değerini gir.
n =np.arange(0,12);                               # n domaini oluştur (periyot: 12 olarak).
x =r*np.exp(1j*Q*n);                              # x işaretini oluştur.

plt.figure(61);                                   # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(n, x);                                   # stem komutu ayrık zamanda çizdirmeyi sağlar.
plt.xlabel('n');                                  # grafiğin x eksenindeki etiketi.
plt.ylabel('x[n]');                               # grafiğin y eksenindeki etiketi.
plt.title('Soru 2.7: a)');                        # grafiğin başlığı.
plt.show();                                       # grafiği göster.

#soruda verilen işaret, 12 örnekleme ile 1 periyot olarak çizdirilmiştir.
```

## KISIM 9)

```python
# 𝑥[𝑛]=𝐴𝛼^n şeklinde olan işaretimiz için;

#a) 1<|𝐴|<10, |𝛼1|<1 ve 𝛼1>0 koşulunu sağlayan değerler seçelim.

A1 =3;                    # rastgele belirlenen A.
a11 =(0.7);               # rastgele belirlenen a1.
n =np.arange(-10,10);     # n'değerlerini istediğimiz gibi belirttik.
x1 =A1*(a11**n);          # x'işaretimizi tanımladık.

plt.figure(91);           # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(n,x1);           # stem ile ayrık zamanlı olarak çizelim.
plt.title('$x[n]=%4.2f (%4.2f)^n$' %(A1, a11));  #başlık.
plt.show();         
```

```python
# işaretimiz '𝑥[𝑛]=𝑠𝑖𝑛(𝜔0𝑛+𝛼)' şeklindedir.
# soru için a =pi/3 seçilmiştir.

#a) 𝜔0=0.1𝜋

a =pi/3;                                 # a değerimizi atayalım.
𝜔0a =0.1*pi;                            # w0 değerimizi girelim.  
n1 =np.arange(0,41);                   # n' domainimizi oluşturalım.
x1 =np.sin(𝜔0a*n1+a);                 # x1 işaretimizi oluşturalım.

plt.figure(101);                       # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(n1,x1);                      # stem ile ayrık zamanda grafiği oluştur.
plt.ylabel('x[n]');                   # y ekseni etiketi.
plt.xlabel('n');                     # x ekseni etiketi.
plt.title('a) \u03C90=0.1\u03C0');            # grafik başlığı.
plt.show();                         # grafiği göster. 

# \u03C90 for small omega
# \u03C0 for pi 
```
