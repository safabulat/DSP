Ders kitabının ikinci bölümünün sonundaki 33. Soruda fark denklemi ile ifade edilen sistemin dürtü
cevabını bilgisayarda bulunuz. Bu sorunun çözümünde dürtü cevabı için hazır formül
kullanılmamalıdır; bu sisteme dürtü girdiğinde sistemin üreteceği cevabı bulmak için kendi
fonksiyonunuzu yazınız.

y[n] = −2x[n] + 4x[n − 1] − 2x[n − 2] verilmiştir.

```python
''' 
dürtü cevabı bulmak istiyorsak x[n]=dirac[n] yazarız,
bu durumda y[n]=h[n] olur. Yukarıda verilen y[n]'e göre
bunu yapacak olursak;

h[n]= -2dirac[n]+4dirac[n-1]-2dirac[n-2] olarak dürtü cevabını buluruz.

tanımlayıp çizecek olursak;
'''
def dirac(n):
    if n == 0:
        return 1;
    else:
        return 0;

n = np.arange(-3,5);
h_ = [];
for i in range(-3,5):
    h = 2*dirac(i)-4*dirac(i-1)-2*dirac(i-2);
    h_.append(h)
''' işaretimizi tanımladık, n'e göre çizdirelim.'''
plt.figure(600);
plt.stem(range(-3,5),h_);
plt.xlabel('n');
plt.ylabel('$h[n]$');
plt.title("Soru 6, h[n] grafiği");
```

## SORU 7)

Ders kitabının ikinci bölümünün sonundaki 36. soruda verilen frekans cevabını bilgisayarda çiziniz ve
elde ettiğiniz grafikten yararlanarak (c) maddesinde verilen problemi çözünüz.

```python
w = np.linspace(0,2*pi);
#omega değerimizi -pi ve pi aralığında aldık 
H = ((1/(1-0.8*np.exp(-1j*w)))+(np.exp(-2*1j*w)/(1-0.8*np.exp(-1j*w))));
#verilen frekans cevabımızı tanımladık
H_abs = np.abs(H)  #abs() komutundan yararlanarak genliğimizi,
H_pha = np.angle(H)   #angle() komutundan yararlanarak fazımızı bulduk.

plt.figure(101);           # yeni figür açtık.
plt.subplot(1, 2, 1);      # subplot ile tek seferde yazdıralım.
plt.plot(w/pi,H_abs);      # omegaya göre genlik çiz.
plt.title('$|H(e^{j\omega})|$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.grid();                # gridi aç.
plt.subplot(1, 2, 2);      # subplot ile tek seferde yazdıralım.
plt.plot(w/pi,H_pha);      # omegaya göre faz çiz.
plt.title('$ \measuredangle H(e^{j\omega})$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.grid();                # gridi aç.

plt.show();                # grafiği göster.
plt.tight_layout();        # başlıklar çakışmasın diye.
```
