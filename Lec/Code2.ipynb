```python
#Tüm problemlerde u[n] kullanacağımız için önceden tanımlayalım.

n = 0;                  # n domainimizin merkezini oluşturalım.

def u(n):               # u[n]'i tanımlayalım.
    if n >= 0:
        return 1;
    else:
        return 0;
    
# Not: burada n domainini range() komutu ile
# min maxını vererek kullanacağım. Bu yüzden
# ilk önce n=0 gibi bir merkez tanımladım.
```

## SORU 6)

#### problem 18 (a), (c) ve (e) numaralı problemi python kullanarak çizdiriniz.

```python
#18.a
print('a.');
h_a = [];                               # boş bir array oluşturalım.
for i in range(n-5, n+5):               # n' domainimizi -5 ve +5 arasında oluşturalım.
    hA = ((1/2)**i)*u(i);               # a şıkkındaki işaretimizi tanımlayalım.
    h_a.append(hA);                     # float barındırdığı için append ile yeni vektöre ekleyelim.
    
plt.figure(601);                        # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-5, n+5), h_a);         # stem komutu ayrık zamanda çizdirmeyi sağlar.
plt.xlabel('n');                        # grafiğin x eksenindeki etiketi.
plt.ylabel('h.a[n]');                   # grafiğin y eksenindeki etiketi.
plt.title('Soru 18: a)');               # grafiğin başlığı.
plt.grid(True);                         # gridi göster.
plt.show();                             # grafiği göster.
```

```python
#18.c
print('c.');
h_c = [];                               # boş bir array oluşturalım.
for i in range(n-5, n+5):               # n' domainimizi -5 ve +5 arasında oluşturalım.
    hC = (1/2)**np.abs(i);              # c şıkkındaki işaretimizi tanımlayalım.
    h_c.append(hC);                     # float barındırdığı için append ile yeni vektöre ekleyelim.

plt.figure(602);                        # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-5, n+5), h_c);         # stem komutu ayrık zamanda çizdirmeyi sağlar.
plt.xlabel('n');                        # grafiğin x eksenindeki etiketi.
plt.ylabel('h.c[n]');                   # grafiğin y eksenindeki etiketi.
plt.title('Soru 18: c)');               # grafiğin başlığı.
plt.grid(True);                         # gridi göster.
plt.show();                             # grafiği göster.
                
```

```python
#18.e
print('e.');
h_e = [];                               # boş bir array oluşturalım.
for i in range(n-5, n+5):               # n' domainimizi -5 ve +5 arasında oluşturalım.
    hE = ((1/3)**i)*u(i)+((3)**i)*u(-(i+1));  # e şıkkındaki işaretimizi tanımlayalım.
    h_e.append(hE);                     # float barındırdığı için append ile yeni vektöre ekleyelim.


plt.figure(603);                        # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-5, n+5), h_e);         # stem komutu ayrık zamanda çizdirmeyi sağlar.
plt.xlabel('n');                        # grafiğin x eksenindeki etiketi.
plt.ylabel('h.e[n]');                   # grafiğin y eksenindeki etiketi.
plt.title('Soru 18: e)');               # grafiğin başlığı.
plt.grid(True);                         # gridi göster.
plt.show();                             # grafiği göster.
```

```python
#19.e
print('e.');
h_e = [];                               # boş bir array oluşturalım.
for i in range(n-13, n+16):             # n' domainimizi -13 ve +16 arasında oluşturalım.
    hE = ((3/4)**np.abs(i))*np.cos(pi*(i/4)+(pi/4));  # e şıkkındaki işaretimizi tanımlayalım.
    h_e.append(hE);                     # float barındırdığı için append ile yeni vektöre ekleyelim.

plt.figure(702);                        # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-13, n+16), h_e);       # stem komutu ile ayrık zamanda -13,16 aralığında çizelim.
plt.xlabel('n');                        # grafiğin x eksenindeki etiketi.
plt.ylabel('h.e[n]');                   # grafiğin y eksenindeki etiketi.
plt.title('Soru 19: e)');               # grafiğin başlığı.
plt.grid(True);                         # gridi göster.
plt.show();                             # grafiği göster.
```

## SORU 8)
#### problem 21 (c) ve (e) numaralı problemi python kullanarak çizdiriniz.

```python
# öncelikle diğer şıklarda kullanmak üzere,
# x[n] ve dirac[n] işaretlerimizi oluşturalım.

def x(n):                   # x[n]'i tanımlayalım.
    if 4 > n >= -1:
        return 1;
    elif n == 4:
        return (1/2);
    else:
        return 0;
    
def dirac(n):               # dirac[n]'i tanımlayalım.
    if n == 0:
        return 1;
    else:
        return 0;

# fonksiyonlarımızı tanımladıktan sonra x[n] işaretimiz istediğimiz gibi olmuş mu
# diye test etmek için yazdıralım.

x_ = [];                              # içine değerleri yazdırabileceğimiz boş array.
for i in range(n-4, n+7):             # n' domainimizi -4 ve +7 arasında oluşturalım.
    a = x(i);                         # a gibi bir isimle x(i) işaretini
    x_.append(a);                     # üstüste toplayarak x[n]'i oluşturalım.
    
plt.figure(81);                       # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-4, n+7),x_);         # Test etmek için x[n]'i çizdirelim.
plt.title('x[n]');                    # grafik başlığı
plt.grid(True);                       # gridi göster.
plt.show();                           # grafiği göster.

# Not: buradaki -4 ve +7 aralığı rastgele* seçilmiştir.
# *sinyalleri düzgün görüntüleyebileceğimiz alan...
```

```python
#21.c
print('c.');

x_c = [];
for i in range(n-4, n+7):             # n' domainimizi -4 ve +7 arasında oluşturalım.
    c = x(2*i);                         # a gibi bir isimle x(i) işaretini
    x_c.append(c);                     # üstüste toplayarak x[n]'i oluşturalım.
    
plt.figure(801);                        # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-4, n+7), x_c);       # stem komutu ile ayrık zamanda -13,16 aralığında çizelim.
plt.xlabel('n');                        # grafiğin x eksenindeki etiketi.
plt.ylabel('x.c[n]');                   # grafiğin y eksenindeki etiketi.
plt.title('Soru 21: c)');               # grafiğin başlığı.
plt.grid(True);                         # gridi göster.
plt.show();                             # grafiği göster.

# Not: 1/2 ve 3/2 değerlerinde de 1 genliğinde bir işaret
# görmem lazım, ancak float sayırları görüntüleyemiyorum.
# sinyal eksik gibi görünüyor ancak normal şartlarda 2 adet
# daha pulse bulunmaktadır (bulunmalıdır).
```

```python
#21.e
print('e.');

x_e = [];
for i in range(n-4, n+7):             # n' domainimizi -4 ve +7 arasında oluşturalım.
    e = x(i-1)*dirac(i-3);            # e gibi bir isimle x(i) işaretini
    x_e.append(e);                    # üstüste toplayarak x[n]'i oluşturalım.

plt.figure(802);                      # grafikleri ayrı tutmak için figüre ile ayır.
plt.stem(range(n-4, n+7), x_e);       # stem komutu ile ayrık zamanda -4,7 aralığında çizelim.
plt.xlabel('n');                      # grafiğin x eksenindeki etiketi.
plt.ylabel('x.e[n]');                 # grafiğin y eksenindeki etiketi.
plt.title('Soru 21: e)');             # grafiğin başlığı.
plt.grid(True);                       # gridi göster.
plt.show();                           # grafiği göster.
```
