### Soru-2

```python
# h_I)
M = 3; # için;
# omega domaini tanımlayalım;
w = np.linspace(-pi,pi,1000);
# Frekans cevabı;
H = (1/(2*M+1))*((sin((w*(2*M)+1)/2))/(sin(w/2)));

H_abs_I = np.abs(H);
H_pha_I = np.angle(H);


plt.figure(201);           # yeni figür açtık.
plt.subplot(2, 1, 1);      # subplot ile tek seferde yazdıralım.
plt.plot(w/pi,H_abs_I);      # omegaya göre genlik çiz.
plt.title('$|H(e^{j\omega})|$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.grid();                # gridi aç.
plt.subplot(2, 1, 2);      # subplot ile tek seferde yazdıralım.
plt.plot(w/pi,H_pha_I/pi);      # omegaya göre faz çiz.
plt.title('$ \measuredangle H(e^{j\omega})$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.grid();                # gridi aç.

plt.show();                # grafiği göster.
plt.tight_layout();        # başlıklar çakışmasın diye.
```

```python
# k)
N = 4;
n = np.arange(0,N,1);
w_discrete = 2*pi*n/N;
x_3 = sin(2*pi*n/4 - pi/8);

X = fft(x_3,N);

X_abs = np.abs(X);
X_pha = np.angle(X);

plt.figure(301);           # yeni figür açtık.
plt.subplot(2, 1, 1);      # subplot ile tek seferde yazdıralım.
plt.stem(w_discrete/pi,X_abs);      # omegaya göre genlik çiz.
plt.title('$|X(e^{j\omega})|$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.grid();                # gridi aç.
plt.subplot(2, 1, 2);      # subplot ile tek seferde yazdıralım.
plt.stem(w_discrete/pi,X_pha/pi);      # omegaya göre faz çiz.
plt.title('$ \measuredangle X(e^{j\omega})$'); # genlik için başlık.
plt.xlabel('$\omega (\\times \pi)$');           # x ekseni ismi.
plt.ylabel('$rad$');          # y ekseni ismi. radyan bırakılmıştır.
plt.grid();                # gridi aç.

plt.show();                # grafiği göster.
plt.tight_layout();        # başlıklar çakışmasın diye.
```
