## SORU 1)

#### a)

```python
Fs = 400
Ts = 1 / Fs
N = 80
n = np.arange(0,2*N)
x = 7 + 11*cos(10*pi*n*Ts) + 9*sin(20*pi*n*Ts - pi/4)

plt.figure(101)
plt.stem(n, x)
plt.title("N = 80 Örnek")
plt.xlabel("n örnek")
plt.ylabel("x[n]")
plt.grid()
plt.show()
```

#### b)

```python
w = np.arange(-pi,pi,2*pi/N)
x_w = np.fft.fftshift(np.fft.fft(x,N)/N)
X_abs = np.abs(x_w)
X_pha = np.angle(x_w)

plt.figure(102)
plt.subplot(2,1,1)
plt.stem(w/pi, X_abs)
plt.title("Genlik")
plt.xlabel("w*pi")
plt.ylabel("X(w)")
plt.grid()
plt.subplot(2,1,2)
plt.stem(w/pi, X_pha)
plt.title("Faz")
plt.xlabel("w*pi")
plt.ylabel("X(w)")
plt.grid()
plt.show()

plt.tight_layout()
```

    Örneklenen işaretin, frekansı F = 10 Hz'dir. Seyrek örnekleme yaptığımız işaretin frekansı ise F_d = 400/20; F_d = 20 Hz'dir. Nyquist teoremine göre 20 >= 2 * 10; 20=20 olmuştur, yani Nyquist hızında örnekleme yapılmıştır; Bundan hareketle işaretimiz geri elde edilebilir.

F = Fs*L/M olduğunu biliyoruz. Buradan hareketle M= 20 ile seyrek örneklenmiş işaretimizi geri kazanmak için sık örnekleme yaparız ve katsayı olarak L = 20 seçeriz. Daha sonrasında sık örneklenmiş işareti interpolasyon filtresinden geçiriz.

Burada iki tip interpolasyon filtremiz olup bunlar; lineer interpolasyon ve ideal AGF'dir. Lineer interpolasyon işlemlerini zaman domaininde yaparken, ideal AGF işlemlerini frekans domaininde gerçekleştiririz.

Pratikte ideal AGF'nin gerçekleştirilmesi mümkün olmadığı için, sinyalimizi geri kazanmak için lineer interpolasyon tekniği kullanırız. h_lin = 1-|n/L| ; |n| < L , 0 ; diğer.
